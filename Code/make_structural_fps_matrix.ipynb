```python
import pandas as pd
from ast import literal_eval
```

```python
table = pd.read_csv('../Data/CellProfiles/structural-fps-matrix.csv')
```

```python
structural_fps = table.loc[:, ['CID', 'morganB_fps']]
structural_fps.set_index('CID', drop=True, inplace=True)
structural_fps['morganB_fps'] = structural_fps['morganB_fps'].apply(lambda x: literal_eval(x))
structural_fps = pd.DataFrame(data=structural_fps['morganB_fps'].to_list(), index=  structural_fps.index)
structural_fps.head()
```

```python
structural_fps.to_csv('../Data/CellProfiles/structural_fps_matrix.csv', sep = '\t')
```
